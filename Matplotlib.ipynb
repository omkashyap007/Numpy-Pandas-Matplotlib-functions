```python
x = n.linspace(1 ,10 , 100) 
fig = mlt.figure()
mlt.plot(n.sin(x) , "-") 
mlt.plot(n.cos(x) ,"--") 
mlt.show()
```

```python
mlt.figure() 

mlt.subplot(2,1,1)
mlt.plot(x, n.sin(x) , "-")

mlt.subplot(2,1,2)
mlt.plot(x , n.cos(x) , "--")
```

```python
# in mlt.subplots( x ) the x determines the number of graphs you want to create .

fig , ax = mlt.subplots(3) 

ax[0].plot(x , n.sin(x)) 
ax[1].plot(x , n.cos(x)) 
ax[2].plot(x , n.tan(x))
```

```python
mlt.plot(x, n.sin(x - 0), color='blue') # specify color by name
mlt.plot(x, n.sin(x - 1), color='g') # short color code (rgbcmyk)
mlt.plot(x, n.sin(x - 2), color='0.75') # Grayscale between 0 and 1
mlt.plot(x, n.sin(x - 3), color='#FFDD44') # Hex code (RRGGBB from 00 to FF)
mlt.plot(x, n.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 and 1
mlt.plot(x, n.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

```python
mlt.plot(x, n.sin(x - 0), color='blue' , linestyle = "dashed") # specify color by name
mlt.plot(x, n.sin(x - 1), color='g' , linestyle = "solid") # short color code (rgbcmyk)
mlt.plot(x, n.sin(x - 2), color='0.75' ,linestyle = "dashdot") # Grayscale between 0 and 1
mlt.plot(x, n.sin(x - 3), color='#FFDD44',linestyle = "dotted") # Hex code (RRGGBB from 00 to FF)
mlt.plot(x, n.sin(x - 4),  "-g") # RGB tuple, values 0 and 1
mlt.plot(x, n.sin(x - 5), "-c"); # all HTML color names supported
```
